## Import Library 

```python
import pandas as pd
import tensorflow as tf
import numpy as np
from IPython.display import display
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
```

# ANN 

## Load Data

```python
data = pd.read_csv("./Processed_data_final_ver2.csv")
X_data = data.drop(["label","label2"],axis=1)
y_data = data["label2"]
```

## Pre-Processing

```python
## Down Sampling (Rest)
import imblearn
X_data, y_data = imblearn.under_sampling.RandomUnderSampler(sampling_strategy='majority').fit_resample(X_data, y_data)
```

```python
y_data[y_data == "four"] = 1
y_data[y_data == "good"] = 2
y_data[y_data == "gun"] = 3
y_data[y_data == "okay"] = 4
y_data[y_data == "one"] = 5
y_data[y_data == "paper"] = 6
y_data[y_data == "rock"] = 7
y_data[y_data == "rest"] = 0
y_data[y_data == "scissor"] = 8
y_data[y_data == "three"] = 9
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.1, stratify=y_data, random_state=1004)
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
## Dataframe으로 변환
X_train_scaled = pd.DataFrame(X_train_scaled)
X_test_scaled = pd.DataFrame(X_test_scaled)
```

## Modeling

### Model 설정 

```python
import tensorflow as tf

model_1 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(600, activation='relu' ,input_shape =(18,)),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(600, activation='relu'),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(600, activation='relu'),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.BatchNormalization(),
#    tf.keras.layers.Dense(600, activation='relu'),
#    tf.keras.layers.BatchNormalization(),
#    tf.keras.layers.Dense(600, activation='relu'),
#    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(10, activation='softmax'),
#    tf.keras.layers.BatchNormalization(),
#    tf.keras.layers.Dense(10, activation='softmax')
])

model_1.compile(optimizer = tf.keras.optimizers.Adam(0.0001),
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
model_1.summary()
```

```python
model_1.fit(X_train_scaled, y_train, epochs=2000, batch_size=2**8, validation_data=(X_test_scaled, y_test))
```

```python
model_1.evaluate(X_test_scaled, y_test)
```

# Machine Learning

### Load Data

```python
data = pd.read_csv("./Processed_data_final_ver2.csv")
X_data = data.drop(["label","label2"],axis=1)
y_data = data["label2"]
```

### Pre-processing

```python
## Down Sampling (Rest)
import imblearn
X_data, y_data = imblearn.under_sampling.RandomUnderSampler(sampling_strategy='majority').fit_resample(X_data, y_data)
```

### Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.1, stratify=y_data, random_state=1004)
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
## Dataframe으로 변환
X_train_scaled = pd.DataFrame(X_train_scaled)
X_test_scaled = pd.DataFrame(X_test_scaled)
```

### SVM

```python
from sklearn.svm import SVC
svm_gau_1 = SVC(kernel='rbf', C=10, gamma=1)
```

```python
print(cross_val_score(svm_gau_1, X_train_scaled, y_train, cv=10, n_jobs=-1))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_1 = RandomForestClassifier(n_estimators = 1000, random_state=0, n_jobs = -1)
```

```python
print(cross_val_score(random_forest_1, X_train_scaled, y_train, cv=10))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
Logistic=LogisticRegression(solver = "saga", n_jobs = -1)
```

```python
print(cross_val_score(Logistic, X_train_scaled, y_train, cv=10, n_jobs=-1))
```
